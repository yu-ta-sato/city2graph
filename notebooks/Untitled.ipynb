```python
# plot by folium
import folium

m = folium.Map([53.4084, -2.9916], zoom_start=13)

for _, row in gdf.iterrows():
    folium.GeoJson(row.geometry).add_to(m)

m
```

```python
# plot by folium

m = folium.Map([53.4084, -2.9916], zoom_start=13)

for _, row in gdf.iterrows():
    folium.CircleMarker(
        location=[row.geometry.y, row.geometry.x],
        radius=3,
        fill=True,
        color='blue',
        fill_color='blue',
        fill_opacity=1.0,
        popup=row['names']
    ).add_to(m)

m
```
